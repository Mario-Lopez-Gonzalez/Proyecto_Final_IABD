```python
# SUMAR 10 A LAS ETIQUETAS DE LOS ARCHIVOS DE ETIQUETAS
import os

# Directorios de entrada y salida
input_folder = r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\Food Computer Vision Project\Food.v2i.yolov11\test\labels"
output_folder = r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\Food Computer Vision Project\Food.v2i.yolov11\test\labels_nuevos"

# Crear la carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Procesar cada archivo de etiquetas
for filename in os.listdir(input_folder):
    if filename.endswith(".txt"):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)
        
        with open(input_path, "r") as f:
            lines = f.readlines()
        
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if parts:  # Verifica que la línea no esté vacía
                parts[0] = str(int(parts[0]) + 10)  # Sumar 10 a la label
                new_lines.append(" ".join(parts) + "\n")
        
        with open(output_path, "w") as f:
            f.writelines(new_lines)

print("Proceso completado. Archivos guardados en 'dataset/nuevos_labels'")
```

```python
# ELIMINAR ARCHIVOS VACÍOS DE IMAGES Y LABELS
import os

# Ruta de las carpetas
images_folder = r'C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\Food Computer Vision Project\Food.v2i.yolov11\train\images'
labels_folder = r'C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\Food Computer Vision Project\Food.v2i.yolov11\train\labels'

# Recorre todos los archivos en la carpeta de etiquetas
for label_file in os.listdir(labels_folder):
    label_path = os.path.join(labels_folder, label_file)
    
    # Verifica si el archivo de texto está vacío
    if os.path.getsize(label_path) == 0:
        # Si el archivo está vacío, encuentra el nombre base del archivo (sin extensión)
        base_name = os.path.splitext(label_file)[0]
        
        # Define el camino del archivo de imagen correspondiente
        image_file = base_name + '.jpg'
        image_path = os.path.join(images_folder, image_file)
        
        # Elimina el archivo de texto y la imagen
        if os.path.exists(label_path):
            os.remove(label_path)
            print(f'Archivo vacío eliminado: {label_path}')
        
        if os.path.exists(image_path):
            os.remove(image_path)
            print(f'Imagen eliminada: {image_path}')
```

```python
import os
import shutil

# Ruta de la carpeta que contiene subcarpetas con imágenes
source_folder = r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\Fotos_Etiquetadas_Clase\Zanahoria\labels\train"  # Reemplázala con la ruta real

# Ruta de destino donde quieres copiar todas las imágenes
destination_folder = r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\labels"  # Reemplázala con la ruta real

# Crear la carpeta de destino si no existe
os.makedirs(destination_folder, exist_ok=True)

# Recorrer todas las subcarpetas y copiar las imágenes
for root, _, files in os.walk(source_folder):
    for file in files:
        if file.lower().endswith((".txt")):
            source_path = os.path.join(root, file)
            destination_path = os.path.join(destination_folder, file)
            
            # Copiar el archivo
            shutil.copy2(source_path, destination_path)

print("Copia de labels completada.")
```

```python
import os
import shutil
import random

# Definir rutas principales
images_folder = r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\images"  # Carpeta donde están las imágenes
labels_folder = r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\labels"  # Carpeta donde están las labels

# Crear carpetas de destino
output_folders = {
    "train": (r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\images_train", 
              r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\labels_train"),
    "test": (r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\images_test", 
             r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\labels_test"),
    "valid": (r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\images_valid", 
              r"C:\Users\Unai Chasco Deza\Desktop\Proyecto_Final\imagenes\labels_valid")
}

# Crear las carpetas de destino si no existen
for img_folder, lbl_folder in output_folders.values():
    os.makedirs(img_folder, exist_ok=True)
    os.makedirs(lbl_folder, exist_ok=True)

# Obtener lista de imágenes y barajar
images = [f for f in os.listdir(images_folder) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
random.shuffle(images)

# Calcular los tamaños de cada conjunto
total_images = len(images)
train_size = int(total_images * 0.80)
test_size = int(total_images * 0.15)
valid_size = total_images - train_size - test_size  # Asegura que sume 100%

# Dividir las imágenes en conjuntos
train_images = images[:train_size]
test_images = images[train_size:train_size + test_size]
valid_images = images[train_size + test_size:]

# Función para copiar archivos
def copy_files(images_list, img_dest, lbl_dest):
    for img in images_list:
        img_path = os.path.join(images_folder, img)
        lbl_path = os.path.join(labels_folder, img.replace(os.path.splitext(img)[1], ".txt"))  # Reemplaza extensión

        # Copiar imagen
        if os.path.exists(img_path):
            shutil.copy2(img_path, os.path.join(img_dest, img))

        # Copiar label si existe
        if os.path.exists(lbl_path):
            shutil.copy2(lbl_path, os.path.join(lbl_dest, os.path.basename(lbl_path)))

# Copiar archivos a sus respectivas carpetas
copy_files(train_images, output_folders["train"][0], output_folders["train"][1])
copy_files(test_images, output_folders["test"][0], output_folders["test"][1])
copy_files(valid_images, output_folders["valid"][0], output_folders["valid"][1])

print("Separación completada correctamente. Archivos copiados en sus respectivas carpetas.")

```
